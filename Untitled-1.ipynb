```python
import pandas as pd
import numpy as np
```

```python

dados = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'

df = pd.read_csv(dados, encoding='latin-1', sep=';')

display(df.head()) #Melhor visualizar as primeiras linhas do dataframe (Se não colocar o display também funciona, mas o display é mais recomendado para notebooks)
```

```python
df_2024 = df.loc[df['ano'] == 2024] #Exibe as linhas do dataframe onde a coluna 'ano' é igual a 2024. O loc é usado para indexação(filtro) baseada em rótulos.
```

```python
df_2024 = df.loc[df['ano'] == 2024]
df_2024 = df_2024.groupby('cisp')['pessoas_desaparecidas'].sum().reset_index()
df_2024

```

```python
df_baixada_2020 = df.loc[(df['ano'] == 2020) & (df['regiao'] == 'Baixada Fluminense')]
df_baixada_2020
```

```python
df_cadaver_2026 = df.loc[(df['ano'] == 2026)]
df_cadaver_2026 = df_cadaver_2026.groupby('cisp')['encontro_cadaver'].sum().reset_index()
df_cadaver_2026
```

```python
dados = np.array([12,15,17,20,22,25,28,30,35,40])
q1 = np.percentile(dados, 25) #Calcula o primeiro quartil (Q1) dos dados usando a função percentile do NumPy.
q2 = np.percentile(dados, 50) 
q3 = np.percentile(dados, 75) 
print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2. Mediana): {q2}')
print(f'Terceiro quartil (Q3): {q3}')
```

```python
array_furto_celular = np.array(df_furto_celular_cisp['furto_celular'])
q1_furto_celular = np.percentile(array_furto_celular, 25)
q2_furto_celular = np.percentile(array_furto_celular, 50)
q3_furto_celular = np.percentile(array_furto_celular, 75)
print(f'Primeiro quartil (Q1): {q1_furto_celular}')
print(f'Segundo quartil (Q2. Mediana): {q2_furto_celular}')
print(f'Terceiro quartil (Q3): {q3_furto_celular}')
```

```python
df_menor_q1_furto_celular = df_furto_celular_cisp.loc[df_furto_celular_cisp
['furto_celular'] < q1_furto_celular]

df_menor_q1_furto_celular = df_menor_q1_furto_celular.sort_values(by='furto_celular', ascending= False)

df_menor_q1_furto_celular
```
