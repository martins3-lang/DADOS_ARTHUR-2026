```python
dados = np.array([12,15,17,20,22,25,28,30,35,40])
q1 = np.percentile(dados, 25) #Calcula o primeiro quartil (Q1) dos dados usando a função percentile do NumPy.
q2 = np.percentile(dados, 50) 
q3 = np.percentile(dados, 75) 
print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2. Mediana): {q2}')
print(f'Terceiro quartil (Q3): {q3}')
```

```python
array_furto_celular = np.array(df_furto_celular_cisp['furto_celular'])

q1_furto_celular = np.percentile(array_furto_celular, 25)
q2_furto_celular = np.percentile(array_furto_celular, 50)
q3_furto_celular = np.percentile(array_furto_celular, 75)
media = np.mean(array_furto_celular)
mediana = np.median(array_furto_celular)
distancia = (media - mediana) / mediana
iqr = q3_furto_celular - q1_furto_celular
limite_superior = q3_furto_celular + (1.5 * iqr)
limite_inferior = q1_furto_celular - (1.5 * iqr)
print(f'Primeiro quartil (Q1): {q1_furto_celular}')
print(f'Segundo quartil (Q2. Mediana): {q2_furto_celular}')
print(f'Terceiro quartil (Q3): {q3_furto_celular}')
print(f'O valor da média é: {media}')
print(f'O valor da mediana é: {mediana}')
print(f'A distância entre média e mediana é de: {distancia}')
print(f'O limite superior é: {limite_superior}') #Calcula o limite superior para identificar outliers.  IQR é a diferença entre o terceiro e o primeiro quartil.
print(f'O limite inferior é: {limite_inferior}')
```

```python
outliers_cisp_furto_celular = df_furto_celular_cisp.loc[df_furto_celular_cisp['furto_celular'] >= limite_superior] #Identifica os outliers, ou seja, os valores de furto_celular que estão acima do limite superior calculado.

outliers_cisp_furto_celular = outliers_cisp_furto_celular.sort_values(by='furto_celular', ascending=False) 

outliers_cisp_furto_celular 
```

```python
df_menor_q1_furto_celular = df_furto_celular_cisp.loc[(df_furto_celular_cisp['furto_celular'] < q1_furto_celular)]

df_menor_q1_furto_celular = df_menor_q1_furto_celular.sort_values(by='furto_celular')

df_menor_q1_furto_celular
```
