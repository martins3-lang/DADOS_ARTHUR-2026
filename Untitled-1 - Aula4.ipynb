```python
import pandas as pd
import numpy as np
```

```python
dados_1sem = 'Preços semestrais - AUTOMOTIVOS_2025.01.csv'
df_1sem = pd.read_csv(dados_1sem, sep=';')
df_1sem.head(3)
```

```python
dados_2sem = 'Preços semestrais - AUTOMOTIVOS_2025.02.csv'
df_2sem = pd.read_csv(dados_2sem, sep=';')
df_2sem.head(3)
```

```python
df_2sem['CNPJ da Revenda'] = df_2sem['CNPJ da Revenda'].astype(str)
```

```python
df_2sem.info
```

```python
df_anp = pd.concat([df_1sem, df_2sem], ignore_index=True)
```

```python
df_anp['Valor de Venda'] = df_anp['Valor de Venda'].str.replace(',', '.').astype(float) 
```

```python
df_gasolina = df_anp.loc[df_anp['Produto'] == 'GASOLINA']
df_etanol = df_anp.loc[df_anp['Produto'] == 'ETANOL']
```

```python
df_media_est_gaso = df_gasolina.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()
df_media_est_gaso = df_media_est_gaso.sort_values(by='Valor de Venda', ascending=False)
df_media_est_gaso.head(5)
```

```python
df_media_est_eta = df_etanol.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()
df_media_est_eta = df_media_est_eta.sort_values(by='Valor de Venda', ascending=False)
df_media_est_eta.head(5)
```

```python
df_media_reg_gaso = df_gasolina.groupby('Regiao - Sigla')['Valor de Venda'].mean().reset_index()
df_media_reg_gaso = df_media_reg_gaso.sort_values(by='Valor de Venda', ascending=False)
df_media_reg_gaso
```

```python
df_media_reg_eta = df_etanol.groupby('Regiao - Sigla')['Valor de Venda'].mean().reset_index()
df_media_reg_eta = df_media_reg_eta.sort_values(by='Valor de Venda', ascending=False)
df_media_reg_eta
```

```python
df_media_reg_gaso = df_gasolina.groupby('Municipio')['Valor de Venda'].mean().reset_index()
df_media_reg_gaso = df_media_reg_gaso.sort_values(by='Valor de Venda', ascending=False)
df_media_reg_gaso.head(10)
```

```python
df_media_reg_gaso = df_gasolina.groupby('Municipio')['Valor de Venda'].mean().reset_index()
df_media_reg_gaso = df_media_reg_gaso.sort_values(by='Valor de Venda')
df_media_reg_gaso.head(10)
```

```python
df_media_reg_eta = df_etanol.groupby('Municipio')['Valor de Venda'].mean().reset_index()
df_media_reg_eta = df_media_reg_eta.sort_values(by='Valor de Venda', ascending=False)
df_media_reg_eta.head(10)
```

```python
df_media_reg_eta = df_etanol.groupby('Municipio')['Valor de Venda'].mean().reset_index()
df_media_reg_eta = df_media_reg_eta.sort_values(by='Valor de Venda')
df_media_reg_eta.head(10)
```
